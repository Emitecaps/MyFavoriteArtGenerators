```python
import matplotlib.pyplot as plt
import random
import matplotlib.cbook as cbook
from matplotlib.patches import FancyArrowPatch, Shadow
import matplotlib.colors as mcolors

fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])
ax.axis('off')
ax.margins(0,0)
ax.set_ylim(0,1)
ax.set_xlim(0,1)

# fig.set_facecolor('r')
set_bg_color(fig, 'c')
add_rect(ax, 'b')
add_circle(ax,'r')
add_line(ax, 'w')




```

```python
def set_bg_color(fig, color):
    fig.set_facecolor(color)
    
def add_rect(ax,color):
    width = random.random()
    height = random.random()
    x = random.random()
    y = random.random()
    
    rect = Rectangle((x,y),width,height,fc=color)
    ax.add_patch(rect)
    
def add_circle(ax, color):
    x = random.random()
    y = random.random()
    radius = random.random()/2
    
    circle = Circle((x,y), radius=radius, fc=color)
    ax.add_patch(circle)
    
def add_line(ax,color):
    x1 = random.random()
    x2 = random.random()
    y1 = random.random()
    y2 = random.random()
    lw = random.random()*10
    
    ax.plot([x1,x2], [y1,y1], color=color, lw=lw)
    
```

```python
def NFT_minter(n_mints, pixels):
    color_names = list(mcolors.CSS4_COLORS)
    colors = mcolors.CSS4_COLORS
    
    min_items = 1
    max_items = 8
    
    for i in range(n_mints):
        fig = plt.figure(figsize=(5,5))
        ax = fig.add_axes([0,0,1,1])
        ax.axis('off')
        ax.margins(0,0)
        ax.set_ylim(0,1)
        ax.set_xlim(0,1)
        
        bg_color = random.sample(color_names,1)[0]
        set_bg_color(fig,colors[bg_color])
        
        for color in random.sample(color_names, random.randint(min_items,max_items)):
            add_rect(ax,colors[color])
        for color in random.sample(color_names, random.randint(min_items,max_items)):
            add_circle(ax,colors[color])
        for color in random.sample(color_names, random.randint(min_items,max_items)):
            add_line(ax,colors[color])
        for color in random.sample(color_names, random.randint(min_items,max_items)):
            add_line(ax,colors[color])
            
        fig.savefig(f'Users/jerilacson/NFT/Simple-Generative-Art', dpi = pixels/5)
        plt.close();
    
```

```python
NFT_minter(100,1000)
```
