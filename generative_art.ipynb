```python
import random
import math
import matplotlib.pyplot as plt
from samila import GenerativeImage, Projection
```

```python
# creates the generative art
def f1(x,y):
    result = random.uniform(-1,1) * x**2 - math.sin(y**2) + abs(y-x)
    return result
```

```python
def f2 (x,y):
    result = random.uniform(-1,1) * (y**3) + 2*x
    return result
```

```python
g = GenerativeImage(f1,f2)
```

```python
g.generate()
```

```python
# seed value of an image
g.seed
```

```python
g.plot(projection = Projection.LAMBERT,color = 'darkgoldenrod4', bgcolor = 'darkred')
# change supported projection, include projection = Projection.POLAR or AITOFF, HAMMER, LAMBERT, MOLLEWEIDE:
# g.plot(projection = Projection.LAMBERT)
# change color:
# g.plot(projection = Projection.LAMBERT, color = 'pink')
# the pink is too light & want to add a black background:
# # g.plot(projection = Projection.LAMBERT, color = 'pink', bgcolor = 'black')
```
